```python
import pandas as pd
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./iris.csv', header=None)
data.head()
```

```python
y = data[4]
X = data.drop(4, axis=1)
```

```python
X.head()
```

```python
encoder = LabelEncoder()
encoder.fit(y)
encoded_y = encoder.transform(y)
dummy_y = to_categorical(encoded_y)
```

```python
#init model
first_model = Sequential()
#add layers
#fully connected layers
first_model.add(Dense(4, activation='relu', name='fc1', input_shape=(4,)))
first_model.add(Dense(3, activation='softmax', name='fc2'))
#compile model with optimizer, loss function and metrics
first_model.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
#describe NN
first_model.summary()
```

```python
second_model = Sequential()
second_model.add(Dense(16, activation='sigmoid', name='fc1', input_shape=(4,)))
second_model.add(Dense(3, activation='softmax', name='fc2'))
second_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
second_model.summary()
```

```python
third_model = Sequential()
third_model.add(Dense(16, activation='sigmoid', name='fc1', input_shape=(4,)))
third_model.add(Dense(3, activation='softmax', name='fc2'))
third_model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])
third_model.summary()
```

```python
history1 = first_model.fit(X, dummy_y, epochs=100, batch_size=5, validation_split=0.1)
```

```python
history2 = second_model.fit(X, dummy_y, epochs=100, batch_size=5, validation_split=0.1)
```

```python
history3 = third_model.fit(X, dummy_y, epochs=100, batch_size=5, validation_split=0.1)
```

```python
hist = history1.history
fig = plt.figure(figsize=(12, 5))
ax = fig.add_subplot(1, 2, 1)
ax.plot(hist['loss'], lw=3)
ax.set_title('Training loss', size=15)
ax.set_xlabel('Epoch', size=15)
ax.tick_params(axis='both', which='major', labelsize=15)

ax = fig.add_subplot(1, 2, 2)
ax.plot(hist['accuracy'], lw=3)
ax.set_title('Training accuracy', size=15)
ax.set_xlabel('Epoch', size=15)
ax.tick_params(axis='both', which='major', labelsize=15)
plt.tight_layout()
#plt.savefig('ch13-cls-learning-curve.pdf')

plt.show()
```

```python
hist = history2.history
fig = plt.figure(figsize=(12, 5))
ax = fig.add_subplot(1, 2, 1)
ax.plot(hist['loss'], lw=3)
ax.set_title('Training loss', size=15)
ax.set_xlabel('Epoch', size=15)
ax.tick_params(axis='both', which='major', labelsize=15)

ax = fig.add_subplot(1, 2, 2)
ax.plot(hist['accuracy'], lw=3)
ax.set_title('Training accuracy', size=15)
ax.set_xlabel('Epoch', size=15)
ax.tick_params(axis='both', which='major', labelsize=15)
plt.tight_layout()
#plt.savefig('ch13-cls-learning-curve.pdf')

plt.show()
```

```python
hist = history3.history
fig = plt.figure(figsize=(12, 5))
ax = fig.add_subplot(1, 2, 1)
ax.plot(hist['loss'], lw=3)
ax.set_title('Training loss', size=15)
ax.set_xlabel('Epoch', size=15)
ax.tick_params(axis='both', which='major', labelsize=15)

ax = fig.add_subplot(1, 2, 2)
ax.plot(hist['accuracy'], lw=3)
ax.set_title('Training accuracy', size=15)
ax.set_xlabel('Epoch', size=15)
ax.tick_params(axis='both', which='major', labelsize=15)
plt.tight_layout()
#plt.savefig('ch13-cls-learning-curve.pdf')

plt.show()
```
